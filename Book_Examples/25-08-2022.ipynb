```python
# Creamos las distribuciones condicionales dependiendo del genero y de las palabras del genero.
import nltk
cfd = nltk.ConditionalFreqDist((genre, word)
                              for genre in brown.categories()
                              for word in brown.words(categories=genre))
```

```python
# Creamos los arreglos con los generos deseados, así como de las palabras deseadas.
genres = ['romance', 'news']
words = ['love', 'president', 'Love', 'President']
```

```python
# Tabulamos las distribuciones condicionales con las condiciones deseadas.
cfd.tabulate(conditions = genres, samples = words)
```
