# Extracción de Información 

Definición de funciones para extracción, visualización y carga de texto.

```python
# Extracción de información de los archivos, con utf-8.
def text_extraction(path):
    with open('./EXCELSIOR_100_files/' + path, encoding='utf-8') as file:
        text = file.read()
    return text
```

```python
# Visualización del texto de los archivos.
def text_visualisation(text):
    print(text)
```

```python
# Creación de nuevos documentos a partir del texto extraido. 
def text_load(path, text = 'No text'):
    with open('./new_text/' + path, 'w', encoding = 'utf-8') as file:
        file.write(text)
```

```python
# Leemos cada uno de los archivos por medio de un for. 
for article_no in range(960401,960718):
    path = 'e' + str(article_no) + '_mod.htm'
    try:
        print(path)
        text = text_extraction(path)
        #text_visualisation(text)
        text_load(path, text)
        print('Se extrajo información exitosamente de:', path)
        print('')
        print('')
    except:
        print('No se pudo Extraer texto de:', path)
        print('')
        print('')
```
