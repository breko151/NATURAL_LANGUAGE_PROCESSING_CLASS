Definición de funciones para extracción, visualización y carga de texto.

```python
# Extracción de información de los archivos, con utf-8.
def text_extraction(path, root = './EXCELSIOR_100_files/', extension = '.htm'):
    with open(root + path + extension, encoding='utf-8') as file:
        text = file.read()
    return text
```

```python
# Visualización del texto de los archivos.
def text_visualisation(text):
    print(text)
```

```python
# Creación de nuevos documentos a partir del texto extraido. 
def text_load(path, root = './new_text/', extension = '.txt', text = 'No text'):
    with open(root + path + extension, 'w', encoding = 'utf-8') as file:
        file.write(text)
```

```python
# Limpieza de las etiquetas
from bs4 import BeautifulSoup
def text_clean(text):
    soup = BeautifulSoup(text, 'lxml')
    return soup.get_text()
```

```python
# Leemos cada uno de los archivos por medio de un for. 
for article_no in range(960401,960718):
    path = 'e' + str(article_no) + '_mod'
    try:
        print(path)
        text = text_extraction(path = path)
        print('Se extrajo información exitosamente de:', path)
        text = text_clean(text)
        print('Se limpio información exitosamente de:', path)
        text_load(path = path, text = text)
        print('Se cargo información de forma correcta')
        print('')
        print('')
    except Exception as e:
        print('Error',e)
        print('')
        print('')
```

Tokenizamos las Palabras del texto

```python
for article_no in range(960401,960718):
    path = 'e' + str(article_no) + '_mod'
    words = []
    try:
        print(path)
        text = text_extraction(path = path, root = './new_text/', extension = '.txt')
        print('Se extrajo información exitosamente de:', path)
        words_1 = text.split()
        print('Tokenización Correcta')
        for word in words_1:
            words.append(word)
        print('Agregando Palabras al Vocabulario')
        print('')
        print('')
    except Exception as e:
        print('Error',e)
        print('')
        print('')
```
