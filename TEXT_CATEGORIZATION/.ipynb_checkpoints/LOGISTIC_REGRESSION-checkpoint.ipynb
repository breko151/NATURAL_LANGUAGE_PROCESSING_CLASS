```python
import numpy as np
import random
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
```

```python
def get_lines(corpus):
    all_lines = list()
    with open(corpus, 'r', encoding = 'latin-1') as rfile:
        for line in rfile:
            all_lines.append(line)
    return all_lines
```

```python
def tokenize(lines):
    new_lines = list()
    for line in all_lines:
        new_line = line.lower()
        new_lines.append(word_tokenize(new_line))
    return new_lines
```

```python
def clean_text(lines):
    new_lines = list()
    for line in lines:
        new_line = list()
        for word in line:
            token = list()
            for c in word:
                #[a-záéíóúñü+$]
                if re.match(r'^[a-záéíóúñü+$]', c):
                    token.append(c)
            token = ''.join(token)
            if token != '':
                new_line.append(token)
        new_lines.append(new_line)
    return new_lines
```

```python
def get_X_y(lines):
    X = list()
    y = list()
    for line in lines:
        n = len(line)
        tag = line.pop(n - 1)
        corpus = line
        X.append(corpus)
        y.append(tag)
    return [X, y]
```

```python
def transform_tag(y):
    new_y = list()
    for i in y:
        if i == 'spam':
            new_i = 1
        else:
            new_i = 0
        new_y.append(new_i)
    return np.array(new_y)
```

```python
def lemmatize_X(X):
    lemmas_X = list()
    wnl = nltk.WordNetLemmatizer()
    for line in X:
        lemmas_line = list()
        for token in line:
            lemmatize_token = wnl.lemmatize(token)
            lemmas_line.append(lemmatize_token)
        lemmas_X.append(lemmas_line)
    return lemmas_X
```

```python
def delete_stopwords(X):
    stopwords_list = stopwords.words('english')
    new_X = list()
    for line in X:
        new_line = list()
        for word in line:
            if word not in stopwords_list:
                new_line.append(word)
        new_X.append(new_line)
    return new_X
```

```python
def get_vocabulary(X):
    words = list()
    for line in X:
        for word in line:
            words.append(word)
    vocabulary = list(sorted(set(words)))
    return vocabulary
```

```python
def get_Corpus(X, vocabulary):
    matrix_X = list()
    for line in X:
        xi = list()
        xi.append(1)
        m = len(line)
        for word in vocabulary:
            if m != 0:
                xi.append(line.count(word) / m) 
            else:
                xi.append(m)
        xi = np.array(xi)
        matrix_X.append(xi)
    matrix_X =  np.array(matrix_X)
    return matrix_X.T
```

```python
def initialize_w(n):
    w = np.zeros(n)
    return w
```

```python
def get_z(Matrix, w):
    z = np.dot(w.T, Matrix)
    return z
```

```python
def prediction(z):
    new_y = np.exp(z) / (1 + np.exp(z))
    return np.array(new_y)
```

```python
def j(y, prediction):
    sum1 = np.sum(y * np.log(prediction))
    sum2 = np.sum((1 - y) * np.log(1 - prediction))
    result = - (sum1 + sum2) / len(y)
    return result
```

```python
def j_partial(X, y, prediction):
    result = list()
    for x in X:
        result = (1 / len(y)) * np.dot((prediction - y), x)
    return np.array(result)
```

```python
def get_new_w(w, jpartial, alpha = 0.0005):
    new_w =  w - (alpha * jpartial)
    return new_w
```

```python
def train(X, y, w, iterations = 1000):
    for i in range(iterations):
        z = get_z(X, w)
        new_y = prediction(z)
        error = j(y, new_y)
        print(error)
        new_w = j_partial(X, y, new_y)
        w = get_new_w(w, new_w)
    return w
```

```python
all_lines = get_lines('./SMS_Spam_Corpus_big.txt')
```

```python
random.shuffle(all_lines)
```

```python
tokenized_lines = tokenize(all_lines)
```

```python
clean_lines = clean_text(tokenized_lines)
```

```python
X, y = get_X_y(clean_lines)
```

```python
y = transform_tag(y)
```

```python
X_lemmatized = lemmatize_X(X)
```

```python
vocabulary = get_vocabulary(X_lemmatized)
```

```python
Matrix_X = get_Corpus(X_lemmatized, vocabulary)
```

```python
n = len(vocabulary)
w = initialize_w(n + 1)
```

```python
w = train(Matrix_X, y, w)
```

```python
z = get_z(Matrix_X, w)
```
