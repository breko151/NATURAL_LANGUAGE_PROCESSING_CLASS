## Distribution of Topics in Documents

```python
def get_freq(topics, article):
    freq = dict()
    for word in topics:
        freq[word] = 0
    for sent in article:
        for word in topics:
            if word in sent:
                freq[word] += 1
    pis = dict()
    sum = 0
    for v in freq.values():
        sum += v
    for k, v in freq.items():
        if sum != 0:
            pis[k] = v / sum
        else:
            pis[k] = 0
    return pis
```

```python
def get_distribution(titles, articles, topics):
    distribution_in_articles = dict()
    for n in range(len(titles)):
        distribution_in_articles[titles[n]] = get_freq(topics, articles[n])
    return distribution_in_articles
```

```python
topics = ['méxico', 'gobierno', 'internet', 'política', 'dólar']
dist_topics = get_distribution(titles, lemmatize_articles, topics)
```
